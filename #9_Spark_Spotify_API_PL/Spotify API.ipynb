# SPOTIFY API

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.oauth2 import SpotifyOAuth
import json
import requests
from time import sleep
```

```python
f = open("creds.json","r")
json_data = f.read()
creds_dict = json.loads(json_data)
f.close()

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=creds_dict["Client_ID"],
                                                           client_secret=creds_dict["Client_Secret"]))
```

```python
results = sp.search(q='b', limit=50)
for idx, track in enumerate(results['tracks']['items']):
    print(idx, track['name'])
```

```python
print("result keys: ",list(results.keys()))
print("tracks keys: ",list(results["tracks"].keys()))
print("items is list: ", type(results["tracks"]["items"])==list, ", with len: ", len(results["tracks"]["items"]))

print("\nitems keys: ",list(results["tracks"]["items"][0].keys()))

print("\nalbum keys: ",list(results["tracks"]["items"][0]["album"].keys()))

print("\nalbum id: ",results["tracks"]["items"][0]["album"]["id"])
print("number of album artists: ",len(results["tracks"]["items"][0]["album"]["artists"]))
album_artists = results["tracks"]["items"][0]["album"]["artists"]
album_artists_list = [artist["name"] for artist in album_artists]
print("album artists: ",album_artists_list)
print("album name: ",results["tracks"]["items"][0]["album"]["name"])
print("album total tracks: ",results["tracks"]["items"][0]["album"]["total_tracks"])
print("album release date: ",results["tracks"]["items"][0]["album"]["release_date"])

print("\ntrack name: ",results["tracks"]["items"][0]["name"])
print("track id: ",results["tracks"]["items"][0]["id"])
track_artists = results["tracks"]["items"][0]["artists"]
track_artists_list = [artist["name"] for artist in album_artists]
print("track artists: ",track_artists_list)
print("track duration: ",results["tracks"]["items"][0]["duration_ms"])
print("track number: ",results["tracks"]["items"][0]["track_number"])
print("track type: ",results["tracks"]["items"][0]["type"])
print("track popularity: ",results["tracks"]["items"][0]["popularity"])
print("track explicity: ",results["tracks"]["items"][0]["explicit"])
```
